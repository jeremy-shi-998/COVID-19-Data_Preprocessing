```python
import requests
import numpy as np
import pandas as pd
from bs4 import BeautifulSoup
from datetime import datetime
```

```python
http = 'https://www.health.nsw.gov.au/news/pages/'
```

```python
Covid_2021 ='https://www.health.nsw.gov.au/news/pages/2021-nsw-health.aspx'
addresses = []
html_text = requests.get(Covid_2021).text
soup = BeautifulSoup(html_text, 'lxml')
link_part = soup.find_all('li',class_ = 'dfwp-item')

for links in link_part:
    link_m = links.find_all('li',class_ = 'dfwp-item')
    for link in link_m:
        if 'statistics' in link.text:
            if '01 April 2021' in link.text:
                break
            else:
                addresses.append(http + link.a['href'])
                print(http + link.a['href'])
                
    else:
        continue
    break
```

```python
Confirmed_cases = []
Deaths = []
Tests = []
Vaccinations = []
Date = []
for link in addresses:
    html_text = requests.get(link).text
    soup = BeautifulSoup(html_text, 'lxml')
    date = soup.find('div', class_ = 'newsdate').text.strip()
    print(date)
    Date.append(datetime.strptime(date, '%d %B %Y').strftime('%Y-%m-%d'))
    
    jobs = soup.find_all('td', class_="moh-rteTableOddCol-6")
    if jobs is not None:
        c = jobs[0].text.replace(',','').replace(' ','')
        d = jobs[1].text.replace(',','').replace(' ','')
        t = jobs[2].text.replace(',','').replace(' ','')
        v = jobs[3].text.replace(',','').replace(' ','')
        if c.isnumeric():
            
            Confirmed_cases.append(int(c))
        else:
            Confirmed_cases.append(None)
            
        if d.isnumeric():
            Deaths.append(int(d))
        else:
            Deaths.append(None)
        if t.isnumeric():
            Tests.append(int(t))
        else:
            Tests.append(None)
        if v.isnumeric():
            Vaccinations.append(int(v))
        else:
            Vaccinations.append(None)
        
        
    else:
        Confirmed_cases.append(None)
        Deaths.append(None)
        Vaccinations.append(None)
        Tests.append(None)
                               
  
                               
    print(d)     
```

```python
result = np.transpose(np.array([Date, Confirmed_cases, Deaths, Tests, Vaccinations]))
print(result)
```

```python
df = pd.DataFrame(result, columns=['Date','Confimed Cases', 'Deaths', 'Tests', 'Vaccinations'])
df
```

```python
df.to_csv('Vaccinations.csv', index=False)
```
